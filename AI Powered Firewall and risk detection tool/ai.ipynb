```python
import requests
import pandas as pd

# ✅ Replace with your actual Etherscan API Key
ETHERSCAN_API_KEY = "PBXE85JBI9MM8CUTB1JBXFUXDQS5UJZ5V7"

def fetch_etherscan_transactions(address):
    """Fetch transaction history from Etherscan API for a given wallet address."""
    url = f"https://api.etherscan.io/api?module=account&action=txlist&address={address}&startblock=0&endblock=99999999&sort=desc&apikey={ETHERSCAN_API_KEY}"
    
    response = requests.get(url)
    data = response.json()
    
    if data["status"] == "1":
        df = pd.DataFrame(data["result"])
        
        # Convert timestamp to readable format
        df["timeStamp"] = pd.to_datetime(df["timeStamp"], unit="s")
        
        # Convert value to Ether from Wei
        df["value"] = df["value"].astype(float) / 10**18
        
        # Select relevant columns
        df = df[["hash", "from", "to", "value", "gasPrice", "gasUsed", "timeStamp"]]
        
        return df
    else:
        print("Error fetching data:", data.get("message", "Unknown error"))
        return None

def find_duplicate_transactions(df):
    """Find duplicate transactions where the same 'from', 'to', and 'value' appear more than once."""
    duplicates = df[df.duplicated(subset=["from", "to", "value"], keep=False)]
    return duplicates

# ✅ User inputs the wallet address
wallet_address = input("Enter Ethereum Wallet Address: ")

# ✅ Fetch transaction history
transactions_df = fetch_etherscan_transactions(wallet_address)

# ✅ Check if transactions exist
if transactions_df is not None and not transactions_df.empty:
    print("\n✅ Transaction History (Latest 5):")
    print(transactions_df.head())  # Show first 5 transactions
    
    # ✅ Find duplicate transactions
    duplicate_txns = find_duplicate_transactions(transactions_df)
    
    if not duplicate_txns.empty:
        print("\n⚠️ Duplicate Transactions Found:")
        print(duplicate_txns)
    else:
        print("\n✅ No duplicate transactions found.")
else:
    print("\n❌ No transactions found for this address.")
```

```python
import requests
import pandas as pd

# ✅ Replace with your actual Etherscan API Key
ETHERSCAN_API_KEY = "PBXE85JBI9MM8CUTB1JBXFUXDQS5UJZ5V7"

# ✅ Define thresholds (adjust based on your needs)
VERY_SMALL_THRESHOLD = 0.001  # Transactions below this (0.001 ETH) are "Very Small"
VERY_BIG_THRESHOLD = 10       # Transactions above this (10 ETH) are "Very Big"

def fetch_etherscan_transactions(address):
    """Fetch transaction history from Etherscan API for a given wallet address."""
    url = f"https://api.etherscan.io/api?module=account&action=txlist&address={address}&startblock=0&endblock=99999999&sort=desc&apikey={ETHERSCAN_API_KEY}"
    
    response = requests.get(url)
    data = response.json()
    
    if data["status"] == "1":
        df = pd.DataFrame(data["result"])
        
        # Convert timestamp to readable format
        df["timeStamp"] = pd.to_datetime(df["timeStamp"], unit="s")
        
        # Convert value from Wei to Ether
        df["value"] = df["value"].astype(float) / 10**18
        
        # Select relevant columns
        df = df[["hash", "from", "to", "value", "gasPrice", "gasUsed", "timeStamp"]]
        
        return df
    else:
        print("Error fetching data:", data.get("message", "Unknown error"))
        return None

def classify_transactions(df):
    """Classify transactions as Very Small, Normal, or Very Big."""
    df["Category"] = "Normal"  # Default category

    # Mark very small transactions
    df.loc[df["value"] < VERY_SMALL_THRESHOLD, "Category"] = "Very Small"

    # Mark very big transactions
    df.loc[df["value"] > VERY_BIG_THRESHOLD, "Category"] = "Very Big"

    return df

def hold_big_transactions(df):
    """Simulate a 'hold' mechanism by flagging Very Big Transactions."""
    big_transactions = df[df["Category"] == "Very Big"]
    
    if not big_transactions.empty:
        print("\n🚨 Alert: Very Big Transactions Found! These transactions are on HOLD! 🚨")
        print(big_transactions)
    else:
        print("\n✅ No Very Big Transactions found.")

    return big_transactions

# ✅ User inputs the wallet address
wallet_address = input("Enter Ethereum Wallet Address: ")

# ✅ Fetch transaction history
transactions_df = fetch_etherscan_transactions(wallet_address)

# ✅ Check if transactions exist
if transactions_df is not None and not transactions_df.empty:
    # ✅ Classify transactions
    transactions_df = classify_transactions(transactions_df)
    
    print("\n✅ Transaction History (Latest 5):")
    print(transactions_df.head())  # Show first 5 transactions
    
    # ✅ Hold very big transactions
    held_transactions = hold_big_transactions(transactions_df)

    # ✅ Save results
    transactions_df.to_csv("classified_transactions.csv", index=False)
    print("\n📁 Transactions saved to 'classified_transactions.csv'")
else:
    print("\n❌ No transactions found for this address.")
```
